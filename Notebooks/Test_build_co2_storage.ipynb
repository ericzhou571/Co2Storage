```python
import geopandas as gpd
import pandas_bokeh
pandas_bokeh.output_notebook()
capacity_list = ['conservative estimate Mt','neutral estimate Mt','optimistic estimate Mt']

# test function
from build_co2_storage import add_storge2buses
```

```python
onshore = gpd.read_file('../data/region/regions_onshore_elec_s_37.geojson')
co2map = gpd.read_file('../data/complete_map_2020_unit_Mt.geojson')
co2map.geometry = co2map.geometry.buffer(0)
```

```python
onshore = onshore.set_index('name')
```

```python
add_storge2buses(onshore,['../data/region/regions_offshore_elec_s_37.geojson'],
                '../data/complete_map_2020_unit_Mt.geojson',
                capacity_name='offshore')
```

```python
import numpy as np
co2map['conservative estimate Mt_original'] = co2map['conservative estimate Mt']
co2map['conservative estimate Mt'].replace(0,np.nan,inplace = True)
co2map['conservative estimate Mt'].fillna(co2map['neutral estimate Mt'],inplace = True)
```

```python
co2map['entire_area'] = co2map.to_crs(crs='EPSG:3395').area
```

```python
df = gpd.clip(co2map,onshore[onshore.name.str.contains('DE')])
co2map.plot_bokeh(
    figsize=(900, 600),
    simplify_shapes=5000,
    dropdown=capacity_list,
    colormap="Viridis",
    hovertool_columns=capacity_list+['ID'],
    colormap_range = (0,100)
    )
```
